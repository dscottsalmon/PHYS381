```python
import matplotlib.pyplot as plt
import numpy as np
import math

#Parameters
gravity = 9.81 #m/s^2
B = 1.6e-4 #Ns/m^2
C = 0.25 #Ns/m^2
diameter = 1e-4 #m
b = B*diameter
c = C*(diameter**2)
theta = float(input("Input value for θ in degrees:")) #takes user input for θ value
speed = float(input("Input value for initial speed in m/s:")) #takes user input for v value
mass = float(input("Input value for mass of sphere in kg:")) #takes user input for mass value
radian = theta*(math.pi/180)
dt = 1e-7
v_x_0 = math.cos(radian) * speed #suppose the sphere was shot at speed at angle theta
v_y_0 = math.sin(radian) * speed #^^^

#initializing variables and lists for vacuum projectile motion
v_y = v_y_0 
time = 0
position_x = 0
position_y = 0
position_x_list_v = [position_x]
position_y_list_v = [position_y]
time_list_v = [time]

i_v = 0 #counter variable used to find peak of motion
while(v_y>0): #while velocity is moving in the positive up direction, keep looping
    time = time + dt #iterate time
    a_y = gravity
    v_y = v_y - (a_y * dt) #find new speed from gravity
    
    position_x = position_x + (v_x_0 * dt) #iterate position in x-axis
    position_y = position_y + (v_y * dt) #iterate position in y-axis
    
    position_x_list_v.append(position_x) #add new info to both vacuum lists
    position_y_list_v.append(position_y)
    time_list_v = [time]
    i_v = i_v + 1 #incremement counter variable by 1

while(position_y >0): #while the sphere is still in the air, but moving in the negative velocity direction
    time = time + dt #iterate time
    
    v_y = v_y + (a_y * dt) #find new speed from gravity
   
    position_x = position_x + (v_x_0 * dt) #iteration position in x-axis
    position_y = position_y - (v_y * dt) #iterate position in y-axis
    
    position_x_list_v.append(position_x) #add new info to both vacuum lists
    position_y_list_v.append(position_y)
    time_list_v = [time]

#initializing variables and lists for linear air resistance projectile motion
time = 0
position_x = 0
position_y = 0
v_x = v_x_0
v_y = v_y_0
position_x_list_l = [position_x]
position_y_list_l = [position_y]
time_list_l = [time]

i_l = 0 #counter variable used to find peak of motion
while(v_y>0): #while velocity is moving in the positive up direction, keep looping
    time = time + dt #iterate time
    
    a_x = (-b/mass) * v_x #finds acceleration in x and y directions
    a_y = -gravity - (b/mass) * v_y 
    v_x = v_x + (a_x * dt)  #changes speeds for x and y accordingly
    v_y = v_y + (a_y * dt)
    
    position_x = position_x + (v_x * dt) #iterates position in x-axis
    position_y = position_y + (v_y * dt) #iterates position in y-axis
    
    position_x_list_l.append(position_x) #add new info to all 3 air resistance lists
    position_y_list_l.append(position_y)
    time_list_l.append(time)
    i_l = i_l + 1 #iterate counter variable

while(position_y>0): #while the sphere is still in the air, but moving in the negative velocity direction
    time = time + dt #iterate time
    
    a_x = (-b/mass) * v_x
    a_y = -gravity - (b/mass) * v_y
    v_x = v_x + (a_x * dt)
    v_y = v_y + (a_y * dt)
    
    position_x = position_x + (v_x * dt) #iterates position in x-axis
    position_y = position_y + (v_y * dt) #iterates position in y-axis
    
    position_x_list_l.append(position_x) #add new info to all 3 air resistance lists
    position_y_list_l.append(position_y)
    time_list_l.append(time)

#initializing lists and variables for the quadratic air resistance projectile motion
time = 0
position_x = 0
position_y = 0
v_x = v_x_0
v_y = v_y_0
position_x_list_q = [position_x]
position_y_list_q = [position_y]
time_list_q = [time]

i_q = 0 #counter variable used to find peak of motion
while(v_y>0): #while velocity is moving in the positive up direction, keep looping
    
    time = time + dt #iterate time
    a_x = (-c/mass) * (v_x**2 + v_y**2)**(1/2) * v_x #finds acceleration in x and y directions
    a_y = -gravity - (c/mass) * (v_x**2 + v_y**2)**(1/2) * v_y
    v_x = v_x + (a_x * dt) #changes speed accordingly
    v_y = v_y + (a_y * dt)
    
    position_x = position_x + (v_x * dt) #iterates position in x-axis
    position_y = position_y + (v_y * dt) #iterates position in y-axis
    
    position_x_list_q.append(position_x) #add new info to all 3 air resistance lists
    position_y_list_q.append(position_y)
    time_list_q.append(time)
    i_q = i_q + 1 #iterate counter variable

while(position_y>0): #while the sphere is still in the air, but moving in the negative velocity direction
    time = time + dt #iterate time
    
    a_x = (-c/mass) * (v_x**2 + v_y**2)**(1/2) * v_x #finds acceleration in x and y directions
    a_y = -gravity - (c/mass) * (v_x**2 + v_y**2)**(1/2) * v_y
    v_x = v_x + (a_x * dt) #changes speed accordingly
    v_y = v_y + (a_y * dt)
    
    position_x = position_x + (v_x * dt) #iterates position in x-axis
    position_y = position_y + (v_y * dt) #iterates position in y-axis
    
    position_x_list_q.append(position_x) #add new info to all 3 air resistance lists
    position_y_list_q.append(position_y)
    time_list_q.append(time)
    
#print statements comparing the three plots of motion
print('\nThe maximum height in the vaccum was: %0.2E m.'%position_y_list_v[i_v])
print('The horizontal distance traveled in the vaccum was: %0.2E m.'%(position_x_list_v[-1]))
print('The time of flight in the vacuum was: %0.2E seconds.\n'%(time_list_v[-1]))
print('\nThe maximum height acheived with linear air resistance was: %0.2E m.'%position_y_list_l[i_l])
print('The horizontal distance traveled with linear air resistance was: %0.2E m.'%(position_x_list_l[-1]))
print('The time of flight with linear air resistance was: %0.2E seconds.\n'%(time_list_l[-1]))
print('\nThe maximum height acheived with quadratic air resistance was: %0.2E m.'%position_y_list_q[i_q])
print('The horizontal distance traveled with quadratic air resistance was: %0.2E m.'%(position_x_list_q[-1]))
print('The time of flight with quadratic air resistance was: %0.2E seconds.\n'%(time_list_q[-1]))

#makes graph
plt.figure()
plt.plot(position_x_list_v, position_y_list_v) #plots vacuum motion
plt.plot(position_x_list_l, position_y_list_l) #plots linear air resistance motion
plt.plot(position_x_list_q, position_y_list_q) #plots quadratic air resistance motion
plt.plot(position_x_list_v[i_v], position_y_list_v[i_v], 'ro') #puts marker at peak of vacuum motion
plt.plot(position_x_list_l[i_l], position_y_list_l[i_l], 'ro') #puts marker at peak of linear air resistance motion
plt.plot(position_x_list_q[i_q], position_y_list_q[i_q], 'ro') #puts marker at peak of quadratic air resistance motion
plt.hlines(y=0,xmin= -0.00005, xmax=position_x_list_v[-1]+0.00005, colors='black', linestyles='--') #makes dotted line at 0
plt.title('y(t) vs. x(t) of a sphere with mass = %0.2E kg,\ndiameter = %0.2E m, shot at %0.1f m/s at θ = %d degrees'%(mass,diameter,speed,theta))
plt.legend(['In a Vacuum','Linear Air Resistance','Quadratic Air Resistance'], loc = 'lower right')
plt.xlabel('x(t) (m)')
plt.ylabel('y(t) (m)')
plt.show()
```
