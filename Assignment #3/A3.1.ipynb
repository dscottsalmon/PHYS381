```python
import matplotlib.pyplot as plt
import numpy as np
import math

#Parameters for air resistance
B = 1.6*(10**-4) #Ns/m^2
C = 0.25 #Ns/m^2

#functions for calculating linear, quadratic, and combined air resistance.
def linear_r(velocity,diameter): #bv = BDv
    result = B * diameter * velocity
    return result

def quadratic_r(velocity,diameter): #cv^2 = CD^2v^2
    result = C * (diameter**2) * (velocity**2)
    return result

def f_v(velocity,diameter): #bv + cv^2
    result = linear_r(velocity,diameter) + quadratic_r(velocity,diameter)
    return result

#making empty lists for later
xaxis = []
linear = []
quadratic = []
total = []

#initializing diameter and velocity variables
D = 1 * (10**-6)
v = 1 * (10**-6)

#loop that runs 600 times, slowly increasing v*D every iteration and recording the linear, quadratic and combined air resistance values
i = 0
while (i<600):
    stor = v*D
    if i == 35: #on 25th iteration, record xvalue for the xaxis on graph 1
        graph1 = stor
    if i == 145: #on 145th iteration, record xvalue for the xaxis on graph2
        graph2 = stor
    xaxis.append(stor)
    stor = linear_r(v,D)
    linear.append(stor)
    stor = quadratic_r(v,D)
    quadratic.append(stor)
    stor = f_v(v,D)
    total.append(stor)
    
    D = D + 0.00025 #iterating diameter and velocity variables
    v = v + 0.00025
    i = i + 1

#plots linear dominant
plt.figure()
plt.plot(xaxis,linear)
plt.plot(xaxis,quadratic)
plt.plot(xaxis,total)
plt.xlim(0,graph1)
plt.ylim(0,(0.150* (10**-7)))
plt.legend(['Linear Air Resistance','Quadratic Air Resistance','Total Air Resistance'], loc = 'upper left')
plt.xlabel('Dv (m^2/s)')
plt.ylabel('Magnitude of Air Resistance (N)')
plt.show()

#plots combined relevant
plt.figure()
plt.plot(xaxis,linear)
plt.plot(xaxis,quadratic)
plt.plot(xaxis,total)
plt.xlim(0,graph2)
plt.ylim(0,(0.6*(10**-6)))
plt.legend(['Linear Air Resistance','Quadratic Air Resistance','Total Air Resistance'], loc = 'upper left')
plt.xlabel('Dv (m^2/s)')
plt.ylabel('Magnitude of Air Resistance (N)')
plt.show()

#plots quadratic dominant
plt.figure()
plt.plot(xaxis,linear)
plt.plot(xaxis,quadratic)
plt.plot(xaxis,total)
plt.xlim(0,0.01)
plt.ylim(0,0.00002)
plt.legend(['Linear Air Resistance','Quadratic Air Resistance','Total Air Resistance'], loc = 'upper left')
plt.xlabel('Dv (m^2/s)')
plt.ylabel('Magnitude of Air Resistance (N)')
plt.show()

#print statements about the range of relevance
print("\n\nFrom the graphs above, it is fair to assume that when Dv < ~7.0 * 10^-5, only linear is relevant.")
print("Likewise, when Dv > ~0.007, or 7.0 * 10^-3, only quadratic is relevant.")
print("\n\nSo a range that both being relevant would approximately be:\n7.0E-5 < Dv < 7.0E-2\n\n")

#Dv calculations for given objects
baseball = 0.07 * 5 #7cm * 5m/s
oil = (1.5e-6) * (5.0e-5) #1.5 * 10^-6m * 5.0 * 10^-5 m/s
raindrop = 0.001 * 1 #0.001m * 1m/s, 1mm = 0.001m

#print statements that detail which kind of air resistance each object experiences
print("The Dv for the baseball, oil drop and raindrop respectively are: %0.3f, %0.2E, and %0.3f (m^2/s)"%(baseball,oil,raindrop))
print("Using these values, and the range determined above, we can conclude that:")
print("\t1. The air resistance experienced by the baseball will be primarily quadratic.")
print("\t2. The air resistance experienced by the oil drop will be primarily linear.")
print("\t3. The air resistance experienced by the raindrop will be a combination of the two.")
```
