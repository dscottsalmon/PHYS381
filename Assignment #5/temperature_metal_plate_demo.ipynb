```python
import numpy as np
import matplotlib.pyplot as plt

# set maximum number of iterations
maxIter = 500

# set dimensions
lenX = lenY = 20
delta = 1

# Boundary conditions
Ttop = 100
Tbottom = 0
Tleft = 0
Tright = 0

# Intitial guess 
Tguess = 30

# set colour interpolation (mapping)
colorinterpolation = 50
colourMap = plt.cm.jet

# set meshgrid
X, Y = np.meshgrid(np.arange(0,lenX), np.arange(0,lenY))

# set an array to store temperature values at each point of the mesh
T = np.empty((lenX, lenY))
T.fill(Tguess)

# set boundary conditions in T array
T[(lenY-1):, :] = Ttop
T[:1, :] = Tbottom
T[:, (lenX-1):] = Tright
T[:, :1] = Tleft

print(T[:, len(X-1):])

# iteration that will solve the Laplace equation for temperature
for iteration in range(0, maxIter):
    for i in range(1,lenX-1, delta):
        for j in range(1, lenY-1, delta):
            T[i,j] = 0.25* (T[i+1,j] + T[i-1,j] + T[i,j+1] + T[i,j-1])
            
            
plt.figure()
plt.contourf(X, Y, T, colorinterpolation, cmap=colourMap)
plt.colorbar()
plt.show()
            
```
