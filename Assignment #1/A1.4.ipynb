```python
import math
import numpy as np
import matplotlib.pyplot as plt

#function that creates parabola
def parabola(x):
    result = a*(x**2) + b*x + c #ax^2 + bx + c
    return result

def derivative(x,a,b):
    result = 2*a*x + b #derivative of ax^2 is 2ax, derivative of bx is b, derivative of c is 0
    return result

def NRmethod(x):
    result = x - (parabola(x)/derivative(x,a,b)) #x - [f(x)/f'(x)]
    return result

#coefficients of parabola
a = 2.0
b = 2.0
c = -2.0
x = -1
tol = 0.0001
nsteps = 0

while math.fabs(parabola(x)) > tol:
    
    x = NRmethod(x)
    nsteps += 1 # nsteps = nsteps + 1

print("\nThe root is approximately at x =", x, "and at this approximation, f(x) =", parabola(x),
      "â‰ˆ 0\nThis calculation required this many iterations to determine:", nsteps, "\n")
print("\nThe blue marker is at f(1), and is the initial point used in the NR method approximation. The red marker is the calculated root.\n")

#plots quadratic
xaxis = np.arange(-5,5, 0.2)
yaxis = parabola(xaxis)
plt.figure()
plt.plot(xaxis,yaxis)
plt.plot(xaxis, 0.0*xaxis, 'k--')
plt.plot(-1, parabola(-1), 'bo') #puts blue marker at initial estimated root
plt.plot(x, parabola(x), 'ro') #puts red marker at root
plt.title('2x^2+2x-6') #title/labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#creating lists
tolList=[]
stepList=[]

#while loop runs multiple iterations of what was done above to determine the relationship between tolerance and # of steps required
while tol<0.5: 
    nsteps = 0 #resets steps every iteration
    x = 1
    while math.fabs(parabola(x)) > tol:
        x = NRmethod(x)
        nsteps += 1 # nsteps = nsteps + 1
    stepList.append(nsteps) #adds number of steps and what the tolerance was for each iteration
    tolList.append(tol)
    tol = tol + 0.0005 #increases tolerance slightly every round

print('\n')

#plots logarithmic graph
xaxis = np.log10(tolList)
yaxis = stepList
plt.figure()
plt.plot(xaxis,yaxis)
plt.xlabel('log(tol)')
plt.ylabel('# of steps')
plt.title('Number of Steps Required to find Root vs. Tolerance Level')
plt.show()
```
