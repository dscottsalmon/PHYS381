```python
import math
import numpy as np
import matplotlib.pyplot as plt

def first_derivative(x): #derivative function, U'(x)
    result = -(A/p)*math.e**(-x/p) + (K/x**2) #derivation outlined in manual
    return result

def second_derivative(x): #second derivative function, U''(x)
    result = (A/(p**2)*math.e**(-x/p) - (2*K/x**3))
    return result

def NRmethod(x):
    result = x - (first_derivative(x) / second_derivative(x)) #x - [U'(x)/U''(x)]
    return result

K = 1.44 #1.44eV
A = 1090.0 #1090eV
p = 0.033 #0.033nm
min = 0.2 #0.2nm
tol = 0.0001 #tolerance value

while(math.fabs(first_derivative(min)) > tol): #doing NRmethod until the root is less than or equal to the tolerance value
    min = NRmethod(min)

minpotential = A*math.exp(-1*min/p) - (K/min) #calculating the potential energy at the determined minimum point

#print statement about what the minimum point was and what its calculated potential energy was
print("The calculated minimum point is at x = %0.10f nm.\nThe calculated minimum potential energy is U(%0.10f) = %0.10f eV.\n "%(min, min, minpotential)) 

#making array 
x = np.arange(0.1,1, 0.01) #making an array of x values from 0.1-1
y = A*np.exp(-1*x/p) - (K/x) #making an array of y values U(x)

plt.figure()
plt.plot(x,y)
plt.title('The Interaction Potential Energy\n as a function of distance between Na+ and Cl-')
plt.xlabel('x (nm)')
plt.ylabel('U(x) (eV)')
plt.plot(min, A*np.exp(-1*min/p) - (K/min), 'ro') #puts red marker at minimum
plt.show()

y = (-(A/p)*np.exp(-1*x/p)+(K/(x**2)))*-1 #making an array of y values -(dU/dx)

plt.figure()
plt.plot(x,y)
plt.title('The Negative Derivative of the Interaction Potential Energy\n as a function of the distance between Na+ and Cl-')
plt.xlabel('x (nm)')
plt.ylabel('-(dU/dx)')
plt.plot(min, A*np.exp(-1*min/p) - (K/min), 'ro') #puts red marker at minimum, 'ro') #puts red marker at minimum
plt.show()
```
