```python
import matplotlib.pyplot as plt
import math

#The seed for the RNG generator was: 800816569666100101
#It gave a value of 55 m/s for the cannonball initial speed and 11 m/s for space invader speed

#Setting all needed parameters
invader_velocity = 11 #m/s
invader_height = 30 #m, constant
invader_x = 150 #m, initial location

gravity = 9.81 #m/s^2, constant
mass = 46/1000 #kg, constant
initial_speed = speed = 55 #m/s
theta = 50 #degrees
radian = theta*(math.pi/180) #radians
diameter = 43/1000 #m, diameter of cannonball
C = 0.25 #Ns/m^2
c = C*(diameter**2)

dt = 0.01 #setting delta-t for every iteration of the loop
v_x_0 = math.cos(radian) * initial_speed #suppose the sphere was shot at speed at angle theta
v_y_0 = math.sin(radian) * initial_speed #^^^

#initializing lists and variables required for the quadratic air resistance projectile motion
time = 0
position_x = 0
position_y = 0
v_x = v_x_0
v_y = v_y_0
invader_x_list = [invader_x]
invader_y_list = [invader_height]
invader_y_velocity = [invader_velocity]
position_x_list = [position_x]
position_y_list = [position_y]
cannonball_speed = [speed]
time_list = [time]

print("When the Cannonball is at the height of the space invader, the locations of the cannonball and space invader are:\n")

#this while loop is describing the motion while the projectile is at or greater then a height of 0.
while(position_y>=0):
    
    time = time + dt #iterate time
    a_x = (-c/mass) * (v_x**2 + v_y**2)**(1/2) * v_x #finds acceleration in x and y directions
    a_y = -gravity - (c/mass) * (v_x**2 + v_y**2)**(1/2) * v_y
    v_x = v_x + (a_x * dt) #changes velocities accordingly
    v_y = v_y + (a_y * dt)
    position_x = position_x + (v_x * dt) #iterates position in x-axis
    position_y = position_y + (v_y * dt) #iterates position in y-axis
    invader_x = invader_x + (invader_velocity * dt * -1) #iterates invader x-location, multiplies by -1 because it is moving in left direction
    speed = (v_x**2 + v_y**2)**(1/2) #this finds the magnitude of the velocities (speed of cannonball)

    if position_y >= 29 and position_y <= 31: #if the position of the cannonball is +- 1 of the height of the space invader, then check the location of both
        distance = abs(position_x-invader_x)
        print("\t%0.2fm and %0.2fm, and the distance on the x-axis between the two is: %0.2fm"%(position_x,invader_x, distance))

    invader_x_list.append(invader_x) #updating all the values in each list
    invader_y_list.append(invader_height)
    invader_y_velocity.append(invader_velocity)
    position_x_list.append(position_x)
    position_y_list.append(position_y)
    cannonball_speed.append(speed)
    time_list.append(time)

print("\nAs can be seen, these x-coordinates are never +- 1.0043m of each other, meaning the cannonball and space invader will never collide.\n\n")

#this makes the trajectory plots for the cannonball and space invader
plt.figure()
plt.plot(position_x_list, position_y_list) #plots quadratic air resistance motion
plt.plot(invader_x_list,invader_y_list)
plt.title('Trajectory of a Cannonball launched in the direction of a moving Space Invader. \nAngle = 50 degrees, Velocity = %im/s, Space Invader Velocity = -%im/s'%(initial_speed, invader_velocity))
plt.legend(['Cannonball Trajectory','Space Invader Trajectory'])
plt.xlabel('x(t) (m)')
plt.ylabel('y(t) (m)')
plt.show()

#this makes the velocity plots for the cannonball and space invader
plt.figure()
plt.plot(time_list, cannonball_speed) #plots quadratic air resistance motion
plt.plot(time_list, invader_y_velocity)
plt.title('Magnitude of the Velocities as a function of Time\nfor both the Cannonball and Space Invader')
plt.legend(['Cannonball Velocity','Space Invader Velocity'])
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.show()
```

First, we can tell qualitatively from the Figure that they do not collide. The trajectories of the two only just barely collide, and they collide at the very end of the cannonballs motion. We can also tell qualitatively that the Cannonball and Space Invader never collided with one another. In the loop depicting the motion in the code, there is an if statement that activates whenever the height of the Cannonball is between 29-31 meters. This is because if the Center of the Space Invader is at exactly 30 meters, and the radius of the Space Invader is 1 meter, then the Cannonball will collide whenever it is within 29-31 meters of height and within +-1 meter of the x-coordinate of the Space Invader. Technically the radius of the Cannonball needs to be accounted for as well, but since its diameter is tiny (43mm), it can be safely neglected in this if statement.
